### 1- Download the Kears+Tensorflow tutorial from Canvas. Import all required modules:

```python
import numpy as np

from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.utils import np_utils

from  keras.wrappers.scikit_learn  import  KerasClassifier
from  sklearn.model_selection  import  GridSearchCV
```

### 2- Import the minst dataset, and split it into testing and training as we saw in the tutorial. Then, reshape each sample into a row vector, and scale it by dividing by 255:

```python
from keras.datasets import mnist
 
# By default, the first 60k of MNIST has been defined as training and the rest as testing set: 
(X_train, y_train), (X_test, y_test) = mnist.load_data()

print(X_train.shape)
print(X_test.shape)
```

```python
# Reshape each image pixels into a row of feature table with 28*28=784 features (each pixel is a feature):

X_train = X_train.reshape(X_train.shape[0], 784)

X_test = X_test.reshape(X_test.shape[0], 784)

print(X_train.shape)
print(X_test.shape)
```

```python
# simply scale the features to the range of [0,1]:

X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
X_train /= 255
X_test /= 255
```

```python
# output label:

print (y_train.shape)
print (y_train[:10])
```

### 3- Perform OneHotEncoding for the label y. So, your label will be a vector of 10 elements for each data sample:

```python
y_train = np_utils.to_categorical(y_train, 10)
y_test = np_utils.to_categorical(y_test, 10)
```

```python
# Label after OneHotEncoding:

print (y_train.shape)
print (y_train[:10,:])
```

### 4- Now, define a function called model_creator. This function will define, create, and compile your neural network model according to your structure, and then return the built model as the output. For the ANN neurons/layers, use the same structure as we had in the tutorial:

```python
def model_creator():

    # define:
    model = Sequential()
    input_size = 784
    hidden_neurons = 100
    out_size = 10

    # design the structure:
    ## Designing the ANN Structure (with 784 inputs, 10 outputs and 100 neuron in a hidden layer):
    # -----------------------------------------
    
    # first layer: input layer
    # Input layer does not do any processing, so no need to define the input layer in this problem.
    # -----------------------------------------
    
    # second layer: hidden layer:
    model.add(Dense(hidden_neurons, input_dim = input_size))  # gets its input from the input layer (784)
    model.add(Activation('sigmoid'))
    # -----------------------------------------
    
    # third layer: output layer:
    model.add(Dense(out_size, input_dim = hidden_neurons))  # gets its input from the hidden layer's output
    model.add(Activation('softmax'))
    
    # compile:
    model.compile(loss='categorical_crossentropy', optimizer='adam',  metrics=['accuracy'])
    
    # return: 
    return model
```

### 5- Fix the random state for reproducibility:   

```python
seed = 2
np.random.seed(seed)
```

### 6- Use KerasClassifier class to wrap your model as an object:

```python
model = KerasClassifier(build_fn = model_creator, verbose=2)
```

### 7- Now, run sklearn GridSearch to find the best batch_size and epochs. 
### Search in this range: batch_size = [30 , 50 , 100 ] , epochs = [10 , 15 , 20].  In your GridSearch, the estimator is the above model,  the scoring should be 'neg_log_loss', and you have to use 10-fold CV: 

```python
batch_size = [30, 50, 100]
epochs = [10, 15, 20]

param_grid = dict(batch_size = batch_size, epochs = epochs)

grid = GridSearchCV(estimator = model, param_grid = param_grid, cv = 10, scoring = 'neg_log_loss')

# fit the grid (start the grid search):
grid.fit(X_train, y_train)

# view the results:
#print(grid.cv_results_)
```

#### Note: The accuracy score and best parameters both change if you reverse the order of epochs and batch_size in your param_grid.

### 8- Based on your results, what is the best batch_size and epochs?  
### Now, test your model with the best batch_size and epochs on the testing set.   grid.best_estimator_.model  gives you the best model found and trained in the grid-search. 
### What is the prediction accuracy on the testing set?

```python
# view the best results corresponding to the best structure:
print(grid.best_score_)
print("The best batch_size and epochs are: ", grid.best_params_)
```

```python
parameters = {'batch_size': [50], 'epochs': [20]}

gs = GridSearchCV(estimator = model, param_grid = parameters, cv = 10, scoring = 'neg_log_loss')

# fit the grid (start the grid search):
gs.fit(X_test, y_test)
```

```python
predict = gs.best_estimator_.model.predict(X_test)
print(predict.shape)
```

```python
pred_accuracy = gs.best_estimator_.model.evaluate(X_test, y_test, verbose = 1)

#print("The loss is: ", pred_accuracy[0])
print("The prediction accuracy on the testing set is: ", pred_accuracy[1])
```
